```python
import json
```

```python
# loading the data
def load_data(filename):
    with open(filename, "r") as f:
        data = json.load(f)
    return data
```

```python
data = load_data("store_data.json")
```

```python
def clean_data(data):
    text_to_num = {"one": 1, "two": 2, "three": 3, "four": 4, "five": 5}

    cleaned_data = []
    unique_users = set()
    
    for user in data:
        raw_rating = str(user["rating"]).strip().lower()

        if(raw_rating in text_to_num):
            raw_rating = text_to_num[raw_rating]
        user["rating"] = raw_rating

        # handling missing values
        raw_age = user.get("age")

        if(raw_age == None):
            user["age"] = None

        if(user["name"].strip() in unique_users):
            continue

        unique_users.add(user["name"])
        cleaned_data.append(user)
        
    return cleaned_data
```

```python
data = clean_data(data)
```

get meaningful insights

```python
def get_insights(data):

    #avg rating
    tot_rating = 0
    for user in data:
        tot_rating += float(user["rating"])
    print(f"avg rating = {tot_rating/len(data)}")

    # percentage of users with poor ratings
    poor_ratings = 0

    for user in data:
        if (float(user["rating"]) < 3):
            poor_ratings += 1

    print(f"% of poor user rating = {(poor_ratings/len(data)) * 100}%")
```

```python
get_insights(data)
```

```python
def get_recommendation(data):
    recommendations = []

    for user in data:
        curr_recom = {}
        curr_recom["name"] = user["name"]

        if(float(user["rating"]) >= 4):
            curr_recom["brand"] = "Apple"
        else:
            curr_recom["brand"] = "samsung"
        recommendations.append(curr_recom)
    return recommendations
```

```python
get_recommendation(data)
```
