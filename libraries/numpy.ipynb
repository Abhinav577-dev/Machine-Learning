```python
import numpy as np
import time
```

```python
arr = np.array([1, 2, 3, 4, 5])
print(arr, type(arr))
```

```python
size = 100_000_000

#time

py_list = list(range(size))
start = time.time()
sq_list = [x**2 for x in py_list]

end = time.time()
print(f"python list time = {end-start} seconds")

# numpy array
np_arr = np.array(py_list)
start = time.time()

#vectorization
sq_array = np_arr **2
end = time.time()
print(f"numpy list time = {end-start} seconds")
```

```python
# memory
import sys

print(f"python list size = {sys.getsizeof(py_list) * len(py_list)} bytes")
print(f"numpy array size = {np_arr.nbytes} bytes")
```
